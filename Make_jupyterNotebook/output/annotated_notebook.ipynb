# Importing Libraries
## Essential Libraries for Data Manipulation and Visualization

```
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```
# Create sample sales data
data = {
    'date': pd.date_range('2024-01-01', periods=12),
    'product': ['A', 'B', 'A', 'C', 'B', 'A', 'C', 'B', 'A', 'C', 'B', 'A'],
    'sales': [100, 150, 120, 200, 180, 110, 220, 190, 130, 210, 170, 140],
    'region': ['North', 'South', 'East', 'West', 'North', 'South', 'East', 'West', 'North', 'South', 'East', 'West']
}
df = pd.DataFrame(data)
```

# Creating Sample Sales Data
## Data Structure
### Defining Dates, Products, Sales, and Regions

```
# Display basic information
print("First few rows:")
print(df.head())
print("\nDataFrame shape:", df.shape)
print("\nColumn info:")
print(df.info())
```

```
# Get statistical summary
print(df.describe())
```

# Displaying Basic Information
## Overview of the DataFrame
### Showing First Rows, Shape, and Column Information

```
# Filter and transform data
df['month'] = df['date'].dt.month
df['quarter'] = df['date'].dt.quarter
north_sales = df[df['region'] == 'North']
print(f"North region sales count: {len(north_sales)}")
```

# Getting Statistical Summary
## Descriptive Statistics
### Summary of Sales Data

```
# Group by product and calculate total sales
product_sales = df.groupby('product')['sales'].sum().sort_values(ascending=False)
print("Total sales by product:")
print(product_sales)
```

```
# Create visualizations
fig, axes = plt.subplots(2, 2, figsize=(12, 8))

# Bar chart: sales by product
product_sales.plot(kind='bar', ax=axes[0, 0], color='steelblue')
axes[0, 0].set_title('Total Sales by Product')
axes[0, 0].set_ylabel('Sales Amount')

# Line chart: sales over time
df.set_index('date')['sales'].plot(ax=axes[0, 1], color='orange')
axes[0, 1].set_title('Sales Over Time')
axes[0, 1].set_ylabel('Sales Amount')

# Pie chart: sales by region
region_sales = df.groupby('region')['sales'].sum()
region_sales.plot(kind='pie', ax=axes[1, 0], autopct='%1.1f%%')
axes[1, 0].set_title('Sales Distribution by Region')

# Scatter plot: sales vs month
axes[1, 1].scatter(df['month'], df['sales'], alpha=0.6, s=100, color='green')
axes[1, 1].set_title('Sales by Month')
axes[1, 1].set_xlabel('Month')
axes[1, 1].set_ylabel('Sales Amount')

plt.tight_layout()
plt.show()
```
